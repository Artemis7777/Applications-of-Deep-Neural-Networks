```python
df['led_vol'] = 0.027 * df['led']##feature engineering
df['motor_vol'] = (2*2*2) * df['motors']
df['gear_vol'] = (1*2*2) * df['gears']
df['volume_parts'] = df['led_vol'] + df['motor_vol'] + df['gear_vol']

df_test['led_vol'] = 0.027 * df_test['led']##feature engineering
df_test['motor_vol'] = (2*2*2) * df_test['motors']
df_test['gear_vol'] = (1*2*2) * df_test['gears']
df_test['volume_parts'] = df_test['led_vol'] + df_test['motor_vol'] + df_test['gear_vol']
```

```python
#Calculate total volume
def calculate_volume(metal,shape,h,l,w):
    shape = np.array([SHAPE.index(x) for x in shape])
    
    vol = np.ndarray(shape=(len(metal)))
    vol[shape==0] = (4.0/3.0)*math.pi*((l[shape==0]/2.0)**3)
    vol[shape==1] =l[shape==1]*w[shape==1]*h[shape==1]
    vol[shape==2]=math.pi*((w[shape==2]/2.0)**2.0)*h[shape==2]
    return vol

df['total_volume'] = calculate_volume(df['metal'],df['shape'],df['height'],df['length'],df['width'])
df_test['total_volume'] = calculate_volume(df_test['metal'],df_test['shape'],df_test['height'],df_test['length'],df_test['width'])
```

```python
#Fill in density
def get_density(metal):
    metal = [METAL.index(x) for x in metal]
    return np.array(DENSE)[metal]

df['density'] = get_density(df['metal'])
df_test['density'] = get_density(df_test['metal'])
```

```python
#Fill in cost
def get_cost(metal):
     metal = [METAL.index(x) for x in metal]
     return np.array(METAL_COST)[metal]

df['metal_cost'] = get_cost(df['metal'])
df_test['metal_cost'] = get_cost(df_test['metal'])
```

```python
#calculate partial volume
df['partial_volume'] = df['total_volume'] - df['volume_parts']
df['weight_est1'] = df['partial_volume']*df['density']
```

```python
# Encode a numeric column as zscores
def encode_numeric_zscore(df, name, mean=None, sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df[name] = (df[name] - mean) / sd
    
encode_numeric_zscore(df, 'metal_cost', mean=None, sd=None)
encode_numeric_zscore(df, 'total_volume', mean=None, sd=None)
encode_numeric_zscore(df, 'volume_parts', mean=None, sd=None)
encode_numeric_zscore(df, 'led', mean=None, sd=None)
encode_numeric_zscore(df, 'gears', mean=None, sd=None)
encode_numeric_zscore(df, 'motors', mean=None, sd=None)
encode_numeric_zscore(df,  'led_vol', mean=None, sd=None)
encode_numeric_zscore(df, 'cost', mean=None, sd=None)
encode_numeric_zscore(df, 'partial_volume', mean=None, sd=None)
encode_numeric_zscore(df, 'weight_est1', mean=None, sd=None)



df.drop('shape',1,inplace=True)
df.drop('metal',1,inplace=True)

```

```python
# Convert a Pandas dataframe to the x,y inputs that TensorFlow needs
def to_xy(df, target):
    result = []
    for x in df.columns:
        if x != target:
            result.append(x)
    # find out the type of the target column.  Is it really this hard? :(
    target_type = df[target].dtypes
    target_type = target_type[0] if hasattr(target_type, '__iter__') else target_type
    # Encode to int for classification, float otherwise. TensorFlow likes 32 bits.
    if target_type in (np.int64, np.int32):
        # Classification
        dummies = pd.get_dummies(df[target])
        return df.as_matrix(result).astype(np.float32), dummies.as_matrix().astype(np.float32)
    else:
        # Regression
        return df.as_matrix(result).astype(np.float32), df.as_matrix([target]).astype(np.float32)
import os
import pandas as pd
from scipy.stats import zscore
from keras.models import Sequential
from keras.layers.core import Dense, Activation
import pandas as pd
import io
import requests
import numpy as np
from sklearn import metrics
import pandas as pd
import io
import requests
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn import metrics
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
x,y=to_xy(df,'weight')
x_train, x_test, y_train, y_test = train_test_split(    
    x, y, test_size=0.25, random_state=45)
model = Sequential()
model.add(Dense(40, input_dim=x.shape[1], activation='relu')) # Hidden 1
model.add(Dense(20, activation='relu')) # Hidden 2
model.add(Dense(1)) # Output
model.compile(loss='mean_squared_error', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, verbose=1, mode='auto')
```
